Removing Outliers from data

```python
import numpy as np

# Assuming 'data' is your dataset with columns containing outliers
# Define a function to remove outliers using IQR
def remove_outliers_iqr(data, threshold=1.5):
    # Calculate the first quartile (Q1) and third quartile (Q3) for each column
    data = data[columns_used]
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    
    # Calculate the IQR for each column
    IQR = Q3 - Q1
    
    # Define the lower and upper bounds to identify outliers
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR
    
    # Create a mask to identify rows with outliers in any column
    outlier_mask = ((data < lower_bound) | (data > upper_bound)).any(axis=1)

    data_imputed = data.copy()
    for column in data.columns:
        median = data[column].median()
        data_imputed.loc[outlier_mask, column] = median
    
    # Remove rows with outliers
    data_cleaned = data[~outlier_mask]
    
    return data_cleaned

# Usage example:
# Assuming 'your_data' is your dataset with columns containing outliers
# df1_remove_outlier = remove_outliers_iqr(df4
```

```python
df1_new = remove_outliers_iqr(df1).merge(df1['dominant_emotion'],left_index=True,right_index=True)
df2_new = remove_outliers_iqr(df2).merge(df2['dominant_emotion'],left_index=True,right_index=True)
df3_new = remove_outliers_iqr(df3).merge(df3['dominant_emotion'],left_index=True,right_index=True)
df4_new = remove_outliers_iqr(df4).merge(df4['dominant_emotion'],left_index=True,right_index=True)
df5_new = remove_outliers_iqr(df5).merge(df5['dominant_emotion'],left_index=True,right_index=True)
df6_new = remove_outliers_iqr(df6).merge(df6['dominant_emotion'],left_index=True,right_index=True)
df7_new = remove_outliers_iqr(df7).merge(df7['dominant_emotion'],left_index=True,right_index=True)
df8_new = remove_outliers_iqr(df8).merge(df8['dominant_emotion'],left_index=True,right_index=True)
df9_new = remove_outliers_iqr(df9).merge(df9['dominant_emotion'],left_index=True,right_index=True)
df10_new = remove_outliers_iqr(df10).merge(df10['dominant_emotion'],left_index=True,right_index=True)
l_new = [df1_new,df2_new,df3_new,df4_new,df5_new,df6_new,df7_new,df8_new,df9_new,df10_new]
```
